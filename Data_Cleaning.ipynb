## Data Cleaning

```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
#  Read in the csv
listings_df = pd.read_csv("Resources/nyc_listings_2022sep07.csv")
listings_df
```

```python
# Find the number of null values 
listings_df.isnull().sum()
```

```python
# Drop columns
listings_df = listings_df.drop(["license"], axis=1)
listings_df
```

```python
# Drop rows with null values
listings_df.dropna()
```

## Data Exploration

```python
# Find the total listings for each neighborhood
borough_listings = listings_df.groupby(["neighbourhood_group"]).count()["neighbourhood"].sort_values(ascending = False)
borough_listings
```

```python
# Find the top 5 neighboorhoods in Manhattan
manhattan_listings = listings_df.loc[listings_df["neighbourhood_group"] == "Manhattan"]
manhattan_neighborhood_listings = manhattan_listings.groupby(["neighbourhood"]).count()["room_type"].sort_values(ascending = False)
manhattan_neighborhood_listings.head()
```

```python
# Find the top 5 neighboorhoods in Brooklyn
brooklyn_listings = listings_df.loc[listings_df["neighbourhood_group"] == "Brooklyn"]
brooklyn_neighborhood_listings = brooklyn_listings.groupby(["neighbourhood"]).count()["room_type"].sort_values(ascending = False)
brooklyn_neighborhood_listings.head()
```

```python
# Find the top 5 neighboorhoods in Queens
queens_listings = listings_df.loc[listings_df["neighbourhood_group"] == "Queens"]
queens_neighborhood_listings = queens_listings.groupby(["neighbourhood"]).count()["room_type"].sort_values(ascending = False)
queens_neighborhood_listings.head()
```

```python
# Find the top 5 neighboorhoods in Bronx
bronx_listings = listings_df.loc[listings_df["neighbourhood_group"] == "Bronx"]
bronx_neighborhood_listings = bronx_listings.groupby(["neighbourhood"]).count()["room_type"].sort_values(ascending = False)
bronx_neighborhood_listings.head()
```

```python
# Find the top 5 neighboorhoods in Staten Island
SI_listings = listings_df.loc[listings_df["neighbourhood_group"] == "Staten Island"]
SI_neighborhood_listings = SI_listings.groupby(["neighbourhood"]).count()["room_type"].sort_values(ascending = False)
SI_neighborhood_listings.head()
```

```python
# Find the total listings for each room type
room_listings = listings_df.groupby(["room_type"]).count()["neighbourhood"].sort_values(ascending = False)
room_listings
```

```python
# Find the hosts with the most listings
host_listings = listings_df.groupby(["host_name"]).count()["room_type"].sort_values(ascending = False)
host_listings.head(10)
```

```python
# Save the cleaned data to a csv
listings_df.to_csv("Resources/clean_nyc_listings.csv", index=False)
```
