## Import & Data Cleaning

```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
#  Read in the csv
listings_df = pd.read_csv("Resources/nyc_listings_2022sep07.csv")
listings_df
```

```python
# Rename columns
listings_df.rename(columns = {"neighbourhood_group":"borough", "neighbourhood":"neighborhood"}, inplace = True)
listings_df
```

```python
# Find the number of null values 
listings_df.isnull().sum()
```

```python
# Drop columns
listings_df = listings_df.drop(["reviews_per_month", "calculated_host_listings_count", "number_of_reviews_ltm", "license"], axis=1)
listings_df
```

```python
# Drop rows with null values
listings_df = listings_df.dropna()
listings_df
```

```python
# Find the hosts with the most listings
host_listings = listings_df.groupby(["host_name"]).count()["room_type"].sort_values(ascending = False)
host_listings.head(10)
```

## Exploratory Data Analysis

```python
# Find the total listings for each neighborhood
borough_listings = listings_df.groupby(["borough"]).count()["neighborhood"].sort_values(ascending = False)
borough_listings
```

```python
# Find the top 5 neighborhoods in Manhattan
manhattan_listings = listings_df.loc[listings_df["borough"] == "Manhattan"]
manhattan_neighborhood_listings = manhattan_listings.groupby(["neighborhood"]).count()["room_type"].sort_values(ascending = False)
manhattan_neighborhood_listings.head()
```

```python
# Find the top 5 neighborhoods in Brooklyn
brooklyn_listings = listings_df.loc[listings_df["borough"] == "Brooklyn"]
brooklyn_neighborhood_listings = brooklyn_listings.groupby(["neighborhood"]).count()["room_type"].sort_values(ascending = False)
brooklyn_neighborhood_listings.head()
```

```python
# Find the top 5 neighborhoods in Queens
queens_listings = listings_df.loc[listings_df["borough"] == "Queens"]
queens_neighborhood_listings = queens_listings.groupby(["neighborhood"]).count()["room_type"].sort_values(ascending = False)
queens_neighborhood_listings.head()
```

```python
# Find the top 5 neighborhoods in Bronx
bronx_listings = listings_df.loc[listings_df["borough"] == "Bronx"]
bronx_neighborhood_listings = bronx_listings.groupby(["neighborhood"]).count()["room_type"].sort_values(ascending = False)
bronx_neighborhood_listings.head()
```

```python
# Find the top 5 neighborhoods in Staten Island
SI_listings = listings_df.loc[listings_df["borough"] == "Staten Island"]
SI_neighborhood_listings = SI_listings.groupby(["neighborhood"]).count()["room_type"].sort_values(ascending = False)
SI_neighborhood_listings.head()
```

```python
# Find the total listings for each room type
room_listings = listings_df.groupby(["room_type"]).count()["neighborhood"].sort_values(ascending = False)
room_listings
```

```python
# Find the hosts with the most listings
host_listings = listings_df.groupby(["host_name"]).count()["room_type"].sort_values(ascending = False)
host_listings.head(10)
```

```python
# Identify the number of short-term listings (under 30 days)
short_term_listings = listings_df[listings_df.minimum_nights<30]
short_term_listings
```

```python
# Identify the number of long-term listings (over 30 days)
long_term_listings = listings_df[listings_df.minimum_nights>29]
long_term_listings
```
